```python
# Ver con mayor defensa y menor vida
df.sort_values(['Defense','HP'], ascending=[0,1])
```

```python
#Crear cambios a los datos
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

#otra forma mas rapida de crear la variable total axis 1 es horizontal
df['Total'] = df.iloc[:,4:10].sum(axis=1)

cols = list(df.columns)
df= df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

```python
# Exportar los datos
df.to_csv('modified.csv', index = False)
```

```python
# Filtrar datos
new_df = df.loc[(df['Type 1'] == 'Steel') & (df['Type 2'] == 'Ground') & (df['HP'] > 70)]

new_df
```
